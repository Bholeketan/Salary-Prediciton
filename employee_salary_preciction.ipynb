```python
%%writefile app.py
# (Paste your full Streamlit code here — the one you provided above)
# prompt: above is a ml model help me build a userinterface using streamlit for the above model

import streamlit as st
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

# Load the dataset
@st.cache_data
def load_data():
    # Assuming the Excel file is available at this path
    try:
        data = pd.read_excel('/content/adult 3.csv.xlsx')
    except FileNotFoundError:
        st.error("Ensure 'adult 3.csv.xlsx' is uploaded to your Colab environment.")
        return None
    return data

data = load_data()

if data is not None:
    st.title("Income Prediction Web App")
    st.write("Predict whether an individual's income is >50K or <=50K based on their characteristics.")

    # Preprocessing steps from the notebook
    data.workclass.replace({'?': "others"}, inplace=True)
    data.occupation.replace({'?': "notlisted"}, inplace=True)
    data = data[data['workclass'] != 'Without-pay']
    data = data[data['workclass'] != 'Never-worked']
    data = data[data['education'] != 'Preschool']
    data = data[data['education'] != '1st-4th']
    data = data[data['education'] != '5th-6th']
    data.drop(columns=["education"], inplace=True)
    data = data[(data['age'] <= 70) & (data['age'] >= 17)]

    # Separate features and target before encoding/scaling
    x = data.drop(columns=['income'])
    y = data['income']

    # Label Encoding for categorical features in the original dataset
    categorical_cols = ['workclass', 'occupation', 'marital-status', 'relationship', 'race', 'gender', 'native-country']
    encoder = LabelEncoder()
    for col in categorical_cols:
        x[col] = encoder.fit_transform(x[col])

    # Scaling numerical features
    # Identify numerical columns before scaling
    numerical_cols = x.select_dtypes(include=['int64', 'float64']).columns.tolist()
    scaler = MinMaxScaler()
    x[numerical_cols] = scaler.fit_transform(x[numerical_cols])


    # Train the models
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=23, stratify=y)

    # Train models (only train once)
    @st.cache_resource
    def train_models(x_train, y_train):
        knn = KNeighborsClassifier()
        knn.fit(x_train, y_train)

        lr = LogisticRegression(max_iter=1000) # Increased max_iter for convergence
        lr.fit(x_train, y_train)

        mlp = MLPClassifier(solver='adam', hidden_layer_sizes=(5, 2), random_state=2, max_iter=2000)
        mlp.fit(x_train, y_train)

        return knn, lr, mlp

    knn, lr, mlp = train_models(x_train, y_train)


    st.sidebar.header("Input Features")

    # Create input widgets for each feature
    age = st.sidebar.slider("Age", int(data['age'].min()), int(data['age'].max()), 30)
    workclass = st.sidebar.selectbox("Workclass", data['workclass'].unique())
    fnlwgt = st.sidebar.number_input("Fnlwgt", value=100000)
    education_num = st.sidebar.number_input("Education-num", value=10)
    marital_status = st.sidebar.selectbox("Marital Status", data['marital-status'].unique())
    occupation = st.sidebar.selectbox("Occupation", data['occupation'].unique())
    relationship = st.sidebar.selectbox("Relationship", data['relationship'].unique())
    race = st.sidebar.selectbox("Race", data['race'].unique())
    gender = st.sidebar.selectbox("Gender", data['gender'].unique())
    capital_gain = st.sidebar.number_input("Capital Gain", value=0)
    capital_loss = st.sidebar.number_input("Capital Loss", value=0)
    hours_per_week = st.sidebar.slider("Hours per Week", int(data['hours-per-week'].min()), int(data['hours-per-week'].max()), 40)
    native_country = st.sidebar.selectbox("Native Country", data['native-country'].unique())

    # Create a dictionary from user inputs
    input_data = {
        'age': age,
        'workclass': workclass,
        'fnlwgt': fnlwgt,
        'educational-num': education_num, # Corrected column name
        'marital-status': marital_status,
        'occupation': occupation,
        'relationship': relationship,
        'race': race,
        'gender': gender,
        'capital-gain': capital_gain,
        'capital-loss': capital_loss,
        'hours-per-week': hours_per_week,
        'native-country': native_country
    }

    # Convert input data to a DataFrame
    input_df = pd.DataFrame([input_data])

    # Preprocess the input data (same steps as training data)
    for col in categorical_cols:
        # Handle potential new categories not seen during training by adding them temporarily
        # This is a workaround; a robust solution would handle unseen data more formally
        input_df[col] = input_df[col].astype(str)
        # Temporarily add unseen values to the encoder's classes
        encoder.classes_ = encoder.classes_.tolist() + list(set(input_df[col]) - set(encoder.classes_))
        input_df[col] = encoder.transform(input_df[col])


    input_df[numerical_cols] = scaler.transform(input_df[numerical_cols])


    st.subheader("Prediction Results")

    # Make predictions using each model
    knn_pred = knn.predict(input_df)[0]
    lr_pred = lr.predict(input_df)[0]
    mlp_pred = mlp.predict(input_df)[0]

    st.write(f"K-Nearest Neighbors Prediction: {knn_pred}")
    st.write(f"Logistic Regression Prediction: {lr_pred}")
    st.write(f"MLP Classifier Prediction: {mlp_pred}")

    st.write("Note: Predictions are either '>50K' or '<=50K'")
```
